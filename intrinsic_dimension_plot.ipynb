```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
sbn.set_style("whitegrid")
sbn.set_context(font_scale=1.25)
plt.rcParams['font.size'] = 16
```

```python
category = "asian" # trans muslim woman
dataset_file_path = f"/home/ubuntu/polytope/toxicity_id_eval/{category}.csv"
```

```python
df = pd.read_csv(dataset_file_path)
#print(df)
x = df['ID']#.head()
for i in range(len(x)):
    print(f'ID for sample {i}: {x[i]} ') #[-5:-1]
```

```python
last_id = []

for row in df['ID']:
    last_id.append(row.split(' ')[-1][:-1])
int_list = [int(x) for x in last_id]
df2 =  df
df2['ID'] = int_list
#print(df2['ID'])
```

```python
df2['Label'] = df2['Label'].replace(0,'Non-Toxic Generation')
df2['Label']= df2['Label'].replace(1, 'Toxic Generation')
max_context_length = df2['Context Length']
max_context_length = max_context_length.max()
```

```python
plt.figure(figsize=(10,10))
print( max_context_length)
sbn.lineplot(data=df2, x="Context Length", y="ID", hue="Label", marker='o',markersize=12 , errorbar=None, linewidth=3)
plt.show()
```
